````python
"""
Approach
- Download compressed cruise archives from provided URLs into a local directory.
- Extract each archive into a structured directory while preserving source grouping.
- Recursively scan extracted files to identify bottle, upcast, and downcast files
  based on filename suffix patterns.
- Parse each file:
  - Use line 3 as the header row.
  - Detect the first numeric row to identify where data begins.
  - Read the data table using whitespace-delimited parsing.
- Clean column names to remove units, symbols, and inconsistent formatting.
- Save one cleaned CSV per source file, grouped by product type.
- Build a station metadata lookup table:
  - Extract Cruise_ID and Station_ID from filenames.
  - Parse header lines into key-value metadata pairs.
  - Normalize date fields where present.
  - Store one row per source file with associated metadata.
- Combine per-file CSVs into three datasets:
  - bottle
  - upcast
  - downcast
- Use the station metadata lookup table to join shared metadata fields into each
  combined dataset using source_file as the key.
- Apply additional data wrangling:
  - Normalize date fields to ISO format (YYYY-MM-DD).
  - Fill missing bottle Latitude, Longitude, Date, and Time values using the
    corresponding upcast file for the same station.
- Sort all final outputs by product_type, Cruise_ID, and Station_ID.
- Write final datasets and station metadata table to the output directory.

Note
- Cruise_ID and Station_ID were extracted from filenames rather than file headers
  to preserve formatting and consistency.  Like zero padding.
- Timezone was not described so ISO datetime with timezone value not added to dataset tables

Assumptions
- File formats match the examples exactly.
- Line 3 contains the column names.
- Data begins at the first numeric row.
- Only bottle, upcast, and downcast files are needed.
- Filenames begin with cruise ID and station number, like ar30-03_021.cbot_s
- Time values are already in HH:MM format and do not need normalization.

"""

import io
import re
import tarfile
import shutil
import urllib.request
from pathlib import Path

import pandas as pd


SOURCE_URLS = [
    "https://www.ncei.noaa.gov/archive/archive-management-system/OAS/bin/prd/jquery/download/252116.1.1.tar.gz",
    "https://www.ncei.noaa.gov/archive/archive-management-system/OAS/bin/prd/jquery/download/0251721.1.1.tar.gz",
    "https://www.ncei.noaa.gov/archive/archive-management-system/OAS/bin/prd/jquery/download/0247461.1.1.tar.gz",
    "https://www.ncei.noaa.gov/archive/archive-management-system/OAS/bin/prd/jquery/download/0252117.1.1.tar.gz",
]

OUTPUT_ROOT = Path("/content/ncei_ctd_pipeline")

PRODUCT_PATTERNS = {
    "bottle": (".cbot_s", ".cbot_so", ".btl"),
    "upcast": (".ucc",),
    "downcast": (".dcc",),
}


def ensure_dir(path):
    path.mkdir(parents=True, exist_ok=True)


def detect_product_type(path):
    name = path.name.lower()
    for product_type, suffixes in PRODUCT_PATTERNS.items():
        if name.endswith(suffixes):
            return product_type
    return None


def extract_ids_from_filename(path):
    stem = path.name.split(".", 1)[0]
    match = re.match(r"(?P<cruise>[^_]+)_(?P<station>\d+)$", stem, flags=re.IGNORECASE)
    if match:
        return match.group("cruise").upper(), match.group("station").zfill(3)
    return None, None


def clean_column_name(name):
    replacements = {
        "(1)": "_1",
        "(2)": "_2",
        "(90)": "_90",
        "(68)": "_68",
        "(ml/l)": "_ml_l",
        "(mL/L)": "_mL_L",
        "(um/kg)": "_um_kg",
        "(db)": "_db",
        "(oC)": "_oC",
        "(mg/m^3)": "_mg_m3",
        "(per)": "_per",
        "(psu)": "_psu",
        "/": "_",
        ".": "",
    }
    name = name.strip()
    for old, new in replacements.items():
        name = name.replace(old, new)
    name = re.sub(r"[^A-Za-z0-9_]+", "_", name)
    name = re.sub(r"_+", "_", name).strip("_")
    return name


def first_data_line_index(lines):
    for i, line in enumerate(lines):
        line = line.strip()
        if line and re.match(r"^[+-]?\d+(?:\.\d+)?\s+", line):
            return i
    return None


def parse_colon_pairs(line):
    pattern = re.compile(
        r"([A-Za-z][A-Za-z0-9 /()_.-]*?):\s*(.*?)(?=(?:\s{2,}[A-Za-z][A-Za-z0-9 /()_.-]*?:)|$)"
    )
    pairs = {}
    for match in pattern.finditer(line):
        key = re.sub(r"[^A-Za-z0-9]+", "_", match.group(1).strip()).strip("_")
        value = match.group(2).strip()
        if key:
            pairs[key] = value
    return pairs


def normalize_iso_date(value):
    if value is None or pd.isna(value):
        return value
    value = str(value).strip()
    if not value:
        return value

    for fmt in ("%m/%d/%Y", "%m/%d/%y", "%Y-%m-%d", "%Y/%m/%d", "%d-%b-%Y", "%d-%b-%y"):
        parsed = pd.to_datetime(value, format=fmt, errors="coerce")
        if not pd.isna(parsed):
            return parsed.strftime("%Y-%m-%d")

    parsed = pd.to_datetime(value, errors="coerce")
    if not pd.isna(parsed):
        return parsed.strftime("%Y-%m-%d")

    return value


def download_and_extract(urls, output_root):
    downloads_dir = output_root / "1_downloads"
    extracts_dir = output_root / "2_extracted_archives"
    ensure_dir(downloads_dir)
    ensure_dir(extracts_dir)

    extracted_roots = []

    for url in urls:
        filename = url.split("/")[-1]
        tar_path = downloads_dir / filename
        extract_name = re.sub(r"\.tar\.gz$", "", filename, flags=re.IGNORECASE)
        extract_dir = extracts_dir / extract_name

        if not tar_path.exists() or tar_path.stat().st_size == 0:
            print(f"Downloading {filename}")
            with urllib.request.urlopen(url) as response, open(tar_path, "wb") as out_f:
                shutil.copyfileobj(response, out_f)
        else:
            print(f"Skipping download: {tar_path}")

        ensure_dir(extract_dir)
        if not any(extract_dir.iterdir()):
            print(f"Extracting to {extract_dir}")
            with tarfile.open(tar_path, "r:gz") as tar:
                tar.extractall(extract_dir)
        else:
            print(f"Skipping extraction: {extract_dir}")

        extracted_roots.append(extract_dir)

    return extracted_roots


def find_source_files(extracted_roots):
    files = []
    for root in extracted_roots:
        for path in root.rglob("*"):
            if path.is_file() and detect_product_type(path):
                files.append(path)
    return sorted(files)


def parse_station_metadata(path, relative_root):
    lines = path.read_text(encoding="utf-8", errors="replace").splitlines()
    data_idx = first_data_line_index(lines)
    header_lines = lines[:data_idx] if data_idx is not None else lines
    cruise_id, station_id = extract_ids_from_filename(path)

    record = {
        "source_file": str(path.relative_to(relative_root)),
        "source_filename": path.name,
        "product_type": detect_product_type(path),
        "Cruise_ID": cruise_id,
        "Station_ID": station_id,
    }

    for line in header_lines:
        for key, value in parse_colon_pairs(line).items():
            key_lower = key.lower()

            if key_lower in {"cruise", "cruise_id", "station", "station_id", "sta", "sta_no"}:
                continue

            if re.match(r"^ar\d+(?:-\d+)?_", key_lower):
                continue

            if value == "":
                continue

            if key_lower == "date":
                value = normalize_iso_date(value)

            record[key] = value

    return record


def parse_data_table(path, relative_root):
    lines = path.read_text(encoding="utf-8", errors="replace").splitlines()
    if len(lines) < 3:
        return pd.DataFrame()

    headers = [clean_column_name(x) for x in re.split(r"\s+", lines[2].strip())]
    data_idx = first_data_line_index(lines)
    if data_idx is None:
        return pd.DataFrame()

    data_text = "\n".join(lines[data_idx:]).strip()
    if not data_text:
        return pd.DataFrame()

    df = pd.read_csv(
        io.StringIO(data_text),
        sep=r"\s+",
        header=None,
        engine="python",
        dtype=str,
    )

    df.columns = headers

    date_cols = [c for c in df.columns if c.lower() == "date"]
    for col in date_cols:
        df[col] = df[col].apply(normalize_iso_date)

    cruise_id, station_id = extract_ids_from_filename(path)

    df["source_filename"] = path.name
    df["source_file"] = str(path.relative_to(relative_root))
    df["product_type"] = detect_product_type(path)
    df["Cruise_ID"] = cruise_id
    df["Station_ID"] = station_id
    return df


def save_per_file_csv(df, product_type, source_filename, output_root):
    out_dir = output_root / "3_per_file_csv" / product_type
    ensure_dir(out_dir)
    safe_name = re.sub(r"[^A-Za-z0-9._-]+", "_", source_filename)
    out_path = out_dir / f"{safe_name}.csv"
    df.to_csv(out_path, index=False)
    return out_path


def combine_product_csvs(product_type, output_root):
    csv_dir = output_root / "3_per_file_csv" / product_type
    csv_files = sorted(csv_dir.glob("*.csv")) if csv_dir.exists() else []
    if not csv_files:
        return pd.DataFrame()
    return pd.concat((pd.read_csv(path, dtype=str) for path in csv_files), ignore_index=True, sort=False)


def fill_bottle_from_upcast(df_bottle, df_upcast):
    if df_bottle.empty or df_upcast.empty:
        return df_bottle

    upcast_cols = [c for c in ["Latitude", "Longitude", "Date", "Time"] if c in df_upcast.columns]
    if not upcast_cols:
        return df_bottle

    upcast_lookup = (
        df_upcast[["source_file"] + upcast_cols]
        .dropna(how="all", subset=upcast_cols)
        .drop_duplicates(subset=["source_file"], keep="first")
        .rename(columns={col: f"{col}_upcast" for col in upcast_cols})
    )

    df_bottle = df_bottle.merge(upcast_lookup, on="source_file", how="left")

    # Bottle files often do not carry complete station position and timestamp values.
    # Fill those fields from the matching upcast file because the upcast represents the
    # same station/cast and provides the needed location and time context for the bottle rows.
    for col in upcast_cols:
        if col not in df_bottle.columns:
            df_bottle[col] = pd.NA
        df_bottle[col] = df_bottle[col].fillna(df_bottle[f"{col}_upcast"])

    df_bottle = df_bottle.drop(columns=[f"{col}_upcast" for col in upcast_cols])
    return df_bottle


def join_with_station_metadata(df, station_metadata_lookup):
    if df.empty:
        return df.copy()

    metadata_cols = [c for c in station_metadata_lookup.columns if c not in df.columns or c == "source_file"]
    return df.merge(
        station_metadata_lookup[metadata_cols],
        on="source_file",
        how="left",
        suffixes=("", "_meta"),
    )


def build_all_datasets(source_urls, output_root):
    ensure_dir(output_root)

    extracted_roots = download_and_extract(source_urls, output_root)
    source_files = find_source_files(extracted_roots)
    if not source_files:
        raise FileNotFoundError("No bottle, upcast, or downcast source files were found.")

    relative_root = output_root / "2_extracted_archives"

    station_metadata_lookup = pd.DataFrame(
        parse_station_metadata(path, relative_root) for path in source_files
    ).sort_values(
        ["product_type", "Cruise_ID", "Station_ID", "source_filename"]
    ).reset_index(drop=True)

    for path in source_files:
        product_type = detect_product_type(path)
        df = parse_data_table(path, relative_root)
        if not df.empty:
            save_per_file_csv(df, product_type, path.name, output_root)

    final_output_dir = output_root / "4_final_output"
    ensure_dir(final_output_dir)

    station_metadata_lookup.to_csv(final_output_dir / "station_metadata_lookup.csv", index=False)

    df_bottle = combine_product_csvs("bottle", output_root)
    df_upcast = combine_product_csvs("upcast", output_root)
    df_downcast = combine_product_csvs("downcast", output_root)

    df_bottle = fill_bottle_from_upcast(df_bottle, df_upcast)

    df_bottle = join_with_station_metadata(df_bottle, station_metadata_lookup)
    df_upcast = join_with_station_metadata(df_upcast, station_metadata_lookup)
    df_downcast = join_with_station_metadata(df_downcast, station_metadata_lookup)

    df_bottle = df_bottle.sort_values(
        ["product_type", "Cruise_ID", "Station_ID"]
    ).reset_index(drop=True)

    df_upcast = df_upcast.sort_values(
        ["product_type", "Cruise_ID", "Station_ID"]
    ).reset_index(drop=True)

    df_downcast = df_downcast.sort_values(
        ["product_type", "Cruise_ID", "Station_ID"]
    ).reset_index(drop=True)

    df_bottle.to_csv(final_output_dir / "df_bottle.csv", index=False)
    df_upcast.to_csv(final_output_dir / "df_upcast.csv", index=False)
    df_downcast.to_csv(final_output_dir / "df_downcast.csv", index=False)

    return station_metadata_lookup, df_bottle, df_upcast, df_downcast


station_metadata_lookup, df_bottle, df_upcast, df_downcast = build_all_datasets(
    SOURCE_URLS,
    OUTPUT_ROOT,
)

final_output_dir = OUTPUT_ROOT / "4_final_output"

print("\nOutput summary:")
print(f"- Station metadata table: {final_output_dir / 'station_metadata_lookup.csv'}")
print(f"- Bottle data: {final_output_dir / 'df_bottle.csv'}")
print(f"- Upcast data: {final_output_dir / 'df_upcast.csv'}")
print(f"- Downcast data: {final_output_dir / 'df_downcast.csv'}")
````
